Essa lista de exercĂ­cios terĂĄ como base a classe `Evento` criada em exercĂ­cios anteriores. Primeiramente criaremos a classe abstrata `EventoABC` com os mĂ©todos de instĂąncia abstratos `__str__(self)` e `isConcluido(self)`, indicando que todos as subclasses que dela herdarem devem implementar esses mĂ©todos.

`EventoABC` tambĂ©m possui os atributos `_titulo` (string) e `_descricao` (string), cujos valores sĂŁo recebidos e inicializados no construtor da classe. Note a convenĂ§ĂŁo de nomenclatura indicando o carĂĄter privado desses atributos.

````python
#### Classe EventoABC
from abc import ABC, abstractmethod

class EventoABC(ABC):
    """
    Classe abstrata base para representaĂ§ĂŁo de eventos.
    Define a interface obrigatĂłria para todas as subclasses.
    """

    def __init__(self, titulo, descricao):
        """
        Inicializa um novo evento com tĂ­tulo e descriĂ§ĂŁo.
        O uso do underscore (_) indica que os atributos sĂŁo protegidos.
        """
        self._titulo = titulo
        self._descricao = descricao

    @abstractmethod
    def __str__(self):
        """
        MĂ©todo abstrato que deve retornar uma representaĂ§ĂŁo em 
        string legĂ­vel do evento.
        """
        pass

    @abstractmethod
    def isConcluido(self):
        """
        MĂ©todo abstrato que deve retornar um booleano indicando 
        se o evento foi finalizado.
        """
        pass
````

````python
#### Classe DataHora
from datetime import datetime, timedelta

class DataHora:
    """
    Classe para suporte ao registro de eventos de calendĂĄrio.
    Gerencia a conversĂŁo de strings para objetos datetime e vice-versa.
    """
    
    # Atributo de classe para padronizaĂ§ĂŁo da formataĂ§ĂŁo
    FORMAT = '%d/%m/%Y, %H:%M'

    @property
    def data_hora(self):
        """
        Getter: Retorna o atributo _data_hora formatado como string.
        """
        if hasattr(self, '_data_hora'):
            return self._data_hora.strftime(self.FORMAT)
        return "Data nĂŁo definida"

    @data_hora.setter
    def data_hora(self, valor_string):
        """
        Setter: Recebe uma string, tenta converter para datetime e armazena.
        LanĂ§a ValueError se o formato for invĂĄlido.
        """
        try:
            self._data_hora = datetime.strptime(valor_string, self.FORMAT)
        except ValueError:
            raise ValueError(f"Formato de data invĂĄlido! Use: {self.FORMAT}")

    def isPassado(self):
        """
        Verifica se a data armazenada Ă© anterior ao momento atual.
        """
        if hasattr(self, '_data_hora'):
            return self._data_hora < datetime.now()
        return False

    def somaDias(self, num_dias):
        """
        Soma uma quantidade de dias Ă  data atual e retorna a string formatada.
        """
        if hasattr(self, '_data_hora'):
            data_hora_somada = self._data_hora + timedelta(days=num_dias)
            return data_hora_somada.strftime(self.FORMAT)
        raise AttributeError("Defina uma data antes de somar dias.")

# --- CĂłdigo de Teste ---

try:
    # instanciando o objeto
    dh = DataHora()

    # definindo a data_hora atravĂ©s da propriedade (chama o setter)
    dh.data_hora = '05/02/2024, 12:30'

    # editando a data_hora atravĂ©s da funĂ§ĂŁo somaDias
    # somaDias retorna uma string, que Ă© passada para o setter da propriedade
    dh.data_hora = dh.somaDias(30)

    # imprimindo a data_hora editada (chama o getter) e se Ă© passado
    print(f"Data Atualizada: {dh.data_hora}")
    print(f"Ă uma data passada? {dh.isPassado()}")

except ValueError as e:
    print(f"Erro: {e}")
````

*  Implementa os mĂ©todos abstratos da superclasse:
    * MĂ©todo `isConcluido()` que invoca o mĂ©todo `isPassado()` de `_data_hora` e retorna o seu resultado.
    * MĂ©todo `__str__` que imprime os atributos do evento na forma `"Evento: _titulo, Data: _data_hora, DescriĂ§ĂŁo: _descricao, Concluido: isConcluido()"`. Note que `isConcluido()` Ă© o mĂ©todo de avaliaĂ§ĂŁo implementado.

````python
#### Classe EventoUnico
class EventoUnico(EventoABC):
    """
    Representa um evento que ocorre em uma Ășnica data e hora especĂ­fica.
    Implementa a interface definida em EventoABC.
    """

    def __init__(self, titulo, descricao, data_hora_str):
        """
        Inicializa o evento chamando o construtor da superclasse e 
        configurando o objeto DataHora interno.
        """
        # Inicializa tĂ­tulo e descriĂ§ĂŁo na superclasse EventoABC
        super().__init__(titulo, descricao)
        
        # Cria a instĂąncia de DataHora e define o valor via propriedade
        self._data_hora = DataHora()
        self._data_hora.data_hora = data_hora_str

    def isConcluido(self):
        """
        Verifica se o evento jĂĄ ocorreu baseando-se no horĂĄrio atual.
        """
        return self._data_hora.isPassado()

    def __str__(self):
        """
        Retorna a representaĂ§ĂŁo formatada do evento Ășnico.
        """
        return (f"Evento: {self._titulo}, "
                f"Data: {self._data_hora.data_hora}, "
                f"DescriĂ§ĂŁo: {self._descricao}, "
                f"Concluido: {self.isConcluido()}")

    def editar_data_hora(self, nova_data_hora_str):
        """
        Altera a data e hora do evento utilizando a propriedade do objeto DataHora.
        """
        self._data_hora.data_hora = nova_data_hora_str

# --- CĂłdigo de Teste ---

# criar evento
evento = EventoUnico('ReuniĂŁo', 'Sala 302, prĂ©dio da esquina', '05/10/2023, 16:30')
print(evento)

# editar data do evento (atravĂ©s da funĂ§ĂŁo que manipula a propriedade interna)
evento.editar_data_hora('05/10/2024, 16:30')
print(evento)
````

#### Q3.
Crie a classe `EventoRecorrente`:
* A classe deve herdar de `EventoABC`.
* Possui como atributo prĂłprio uma lista privada de objetos `DataHora` (como vocĂȘ deve nomear o atributo?).
* Seu construtor recebe os atributos da superclasse, alĂ©m dos atributos `data_hora_inicial` (string formatada), `data_hora_final` (string formatada) e `intervalo_repeticao` (int), sendo o intervalo dado em dias. Preencha a coleĂ§ĂŁo `DataHora` de acordo com o intervalo de repetiĂ§ĂŁo fornecido. Dica: crie o objeto `DataHora` inicial e use sua funĂ§ĂŁo interna `somaDias` para criar iterativamente as novas instĂąncias do intervalo atĂ© chegar em `DataHora` final. 
*  Implementa os mĂ©todos abstratos da superclasse:
    * MĂ©todo `isConcluido(indice)` que que invoca o mĂ©todo `isPassado()` do elemento `indice` da coleĂ§ĂŁo de objetos `DataHora` e retorna seu resultado. 
    * MĂ©todo `__str__` que imprime (em um laĂ§o) **todos as ocorrĂȘncias `i` do evento** na forma `"Evento: _titulo, Data: data_hora[i], DescriĂ§ĂŁo: _descricao, Concluido: isConcluido(i)"`. 
* Crie o mĂ©todo `editar_data_hora` que recebe `data_hora_antiga` e `data_hora_nova` e altera o elemento da coleĂ§ĂŁo de objetos `DataHora` que corresponde a `data_hora_antiga` fornecida.    


Teste a classe `EventoRecorrente` com o seguinte cĂłdigo:

```python
    'ReuniĂŁo', 'Sala 302, prĂ©dio da esquina', 
    '05/01/2024, 16:30', '05/01/2025, 16:30', 30)

# imprimir eventos
print(eventos)

# editar um dos eventos
eventos.editar_data_hora('05/12/2024, 16:30', '05/12/2024, 11:30')

# imprimir eventos
print(eventos)
```

````python
from datetime import datetime

class EventoRecorrente(EventoABC):
    """
    Representa um evento que se repete em intervalos regulares.
    MantĂ©m uma coleĂ§ĂŁo de objetos DataHora para cada ocorrĂȘncia.
    """

    def __init__(self, titulo, descricao, data_hora_inicial, data_hora_final, intervalo_repeticao):
        """
        Construtor que gera todas as ocorrĂȘncias do evento entre a data inicial e final.
        """
        super().__init__(titulo, descricao)
        self._datas_horas = []  # Atributo privado (coleĂ§ĂŁo de objetos DataHora)

        # ConfiguraĂ§ĂŁo da data inicial e final para controle do laĂ§o
        dh_atual = DataHora()
        dh_atual.data_hora = data_hora_inicial
        
        # Objeto auxiliar para comparaĂ§ĂŁo de limite
        limite = datetime.strptime(data_hora_final, DataHora.FORMAT)

        # Preenchimento iterativo da coleĂ§ĂŁo
        while datetime.strptime(dh_atual.data_hora, DataHora.FORMAT) <= limite:
            # Criamos uma nova instĂąncia para cada ocorrĂȘncia
            nova_ocorrencia = DataHora()
            nova_ocorrencia.data_hora = dh_atual.data_hora
            self._datas_horas.append(nova_ocorrencia)
            
            # AvanĂ§a a data para a prĂłxima repetiĂ§ĂŁo
            proxima_data_str = dh_atual.somaDias(intervalo_repeticao)
            dh_atual.data_hora = proxima_data_str

    def isConcluido(self, indice):
        """
        Verifica se uma ocorrĂȘncia especĂ­fica (pelo Ă­ndice) jĂĄ passou.
        """
        return self._datas_horas[indice].isPassado()

    def __str__(self):
        """
        Retorna a representaĂ§ĂŁo de todas as ocorrĂȘncias do evento.
        """
        saida = []
        for i, dh in enumerate(self._datas_horas):
            status = "Sim" if self.isConcluido(i) else "NĂŁo"
            saida.append(
                f"Evento: {self._titulo}, Data: {dh.data_hora}, "
                f"DescriĂ§ĂŁo: {self._descricao}, Concluido: {status}"
            )
        return "\n".join(saida)

    def editar_data_hora(self, data_hora_antiga, data_hora_nova):
        """
        Busca uma data especĂ­fica na coleĂ§ĂŁo e a altera para o novo valor.
        """
        for dh in self._datas_horas:
            if dh.data_hora == data_hora_antiga:
                dh.data_hora = data_hora_nova
                break

# --- CĂłdigo de Teste ---

# criar evento recorrente a cada 30 dias
eventos = EventoRecorrente(
    'ReuniĂŁo', 'Sala 302, prĂ©dio da esquina', 
    '05/01/2024, 16:30', '05/01/2025, 16:30', 30)

# imprimir eventos
print("--- Lista de Eventos ---")
print(eventos)

# editar um dos eventos (alterando o horĂĄrio da reuniĂŁo de dezembro)
eventos.editar_data_hora('05/12/2024, 16:30', '05/12/2024, 11:30')

# imprimir eventos novamente para verificar alteraĂ§ĂŁo
print("\n--- Lista ApĂłs EdiĂ§ĂŁo ---")
print(eventos)
````

Por fim, vamos sĂł ver o polimorfismo em aĂ§ĂŁo. Crie e preencha uma lista de eventos, sendo alguns do tipo `EventoUnico` e outros do tipo `EventoRecorrente`. Sobre essa lista, execute o laĂ§o de impressĂŁo a seguir:
```python

```
A funĂ§ĂŁo `print` irĂĄ invocar o mĂ©todo especial `__str__` das classes correspondentes dependendo do tipo do objeto recebido. AĂ­ estĂĄ o polimorfismo :)

````python
# 1. Criando a lista de eventos (o "contĂȘiner" polimĂłrfico)
lista_eventos = []

# 2. Adicionando um Evento Ănico
lista_eventos.append(
    EventoUnico("Entrega de RelatĂłrio", "Enviar PDF para a diretoria da Pattrol", "10/05/2026, 09:00")
)

# 3. Adicionando um Evento Recorrente
lista_eventos.append(
    EventoRecorrente("ReuniĂŁo de Sprints", "SincronizaĂ§ĂŁo da equipe de Engenharia Digital", 
                     "15/05/2026, 14:00", "15/07/2026, 14:00", 30)
)

# 4. Adicionando outro Evento Ănico
lista_eventos.append(
    EventoUnico("ManutenĂ§ĂŁo GeoCraft", "Troca de baterias do Horus 2", "20/05/2026, 10:30")
)

# 5. O LaĂ§o PolimĂłrfico
print(f"{'='*20} CRONOGRAMA DE EVENTOS {'='*20}\n")

for evento in lista_eventos:
    # Aqui o Python nĂŁo sabe se 'evento' Ă© Unico ou Recorrente.
    # Ele apenas chama __str__ e o objeto se vira para se auto-imprimir.
    print(evento)
    print("-" * 60)
````
