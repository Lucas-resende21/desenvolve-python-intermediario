#### Q3. 
Crie um dataframe `participacao` com as colunas `país`, `copa`, e `num_copas` que armazena a quantidade de copas do mundo que cada país participou, informando se é da copa masculina ou feminina. Ao final imprima, como apresentado a seguir, o top 5 países de cada competição que mais participou de copas do mundo.

Exemplo de saída (valores inventados):
```
país            copa        num_copas
Brazil          Feminina    8
Unites States   Feminina    8
Germany         Feminina    8
Japan           Feminina    7
Colombia        Feminina    7
```

```
país            copa        num_copas
Brazil          Masculina   21
Germany         Masculina   21
Argentina       Masculina   20
England         Masculina   20
Mexico          Masculina   20
```

````python
import pandas as pd

# 1. Identificar se a copa é Masculina ou Feminina (baseado na lógica da Q3 anterior)
wc['tipo_copa'] = wc['copa'].apply(lambda x: 'Feminina' if 'Women' in x else 'Masculina')

# 2. Criar uma lista de participações (País + Ano + Tipo de Copa)
# Precisamos olhar tanto para o time_1 quanto para o time_2
part1 = wc[['time_1', 'ano', 'tipo_copa']].rename(columns={'time_1': 'país'})
part2 = wc[['time_2', 'ano', 'tipo_copa']].rename(columns={'time_2': 'país'})

# 3. Concatenar e remover duplicatas (um país participa de vários jogos, mas de 1 copa por ano)
todas_participacoes = pd.concat([part1, part2]).drop_duplicates()

# 4. Agrupar para contar o número de copas por país e tipo
participacao = todas_participacoes.groupby(['país', 'tipo_copa']).size().reset_index(name='num_copas')

# Renomear coluna para o padrão solicitado
participacao = participacao.rename(columns={'tipo_copa': 'copa'})

# 5. Filtrar e imprimir o Top 5 de cada categoria
top_5_feminina = participacao[participacao['copa'] == 'Feminina'].sort_values(by='num_copas', ascending=False).head(5)
top_5_masculina = participacao[participacao['copa'] == 'Masculina'].sort_values(by='num_copas', ascending=False).head(5)

print(top_5_feminina.to_string(index=False))
print("\n")
print(top_5_masculina.to_string(index=False))
````

#### Q4. 
* Crie um dataframe `gols` com duas colunas `país` e `total_gols` com o total de gols marcados por cada país durante todas as copas do mundo, juntando gols em casa (`gols_1`) e gols como time visitante (`gols_2`).
* Imprima o dataframe `gols` ordenado de forma decrescente, para que os times com mais gols fiquem no topo.

Segue um exemplo ilustrativo com o formato do dataframe resultado:

```
país        total_gols
Brazil      120
Argentina   112
Germany     107
...
```

````python
import pandas as pd

# 1. Agrupar gols marcados como time_1
gols_casa = wc.groupby('time_1')['gols_1'].sum().reset_index()
gols_casa.columns = ['país', 'gols']

# 2. Agrupar gols marcados como time_2
gols_fora = wc.groupby('time_2')['gols_2'].sum().reset_index()
gols_fora.columns = ['país', 'gols']

# 3. Concatenar os dois resultados e somar por país
gols = pd.concat([gols_casa, gols_fora]).groupby('país')['gols'].sum().reset_index()

# 4. Renomear para o padrão solicitado e ordenar
gols.columns = ['país', 'total_gols']
gols = gols.sort_values(by='total_gols', ascending=False)

# 5. Imprimir o resultado
print(gols.to_string(index=False))
````

#### Q5. 
Qual país tomou mais cartões amarelos somando todas as copas?

Neste exercício você vai trabalhar com as colunas `cartao_amarelo_1` e `cartao_amarelo_2` onde cada observação é uma string represetando uma lista dos cartões amarelos de um único jogo na forma `minuto|placar|jogador(a)`. Por exemplo:
```
['16’|1:0|Rosana Gómez', '20’|2:0|Gabriela Chávez]
```

Recomendo criar colunas `num_cartoes_1` e `num_cartoes_2` a partir de cada coluna `cartao_amarelo_X` usando o método genérico `apply` para chamar uma função que você vai criar para transformar uma observação de cartão amarelo em uma contagem de elementos da lista. 

Lembre de levar em consideração que muitas observações são `NaN`. 

Em seguida faça sua mágica para agrupar as informações por país, acumular os cartões de jogos em casa e jogos visitante, e produzir o resultado final como apresentado a seguir (valores inventados):

```
país        cartões amarelos
Argentina   97
England     93
Australia   93
...
```

````python
import pandas as pd
import ast

# 1. Função para contar os cartões dentro da string
def contar_cartoes(valor):
    # Se for NaN ou nulo, não há cartões
    if pd.isna(valor) or valor == "" or valor == "[]":
        return 0
    
    # Se o dado vier como string representando uma lista: "['evento1', 'evento2']"
    # Transformamos a string em uma lista real do Python e contamos o tamanho
    try:
        # ast.literal_eval é mais seguro que eval para converter strings em objetos Python
        lista_eventos = ast.literal_eval(valor)
        return len(lista_eventos)
    except:
        # Caso o formato não seja uma lista válida, mas haja texto, 
        # tratamos como 1 cartão ou aplicamos lógica de contagem de separadores
        return 0

# 2. Criar as colunas numéricas de contagem
wc['num_cartoes_1'] = wc['cartao_amarelo_1'].apply(contar_cartoes)
wc['num_cartoes_2'] = wc['cartao_amarelo_2'].apply(contar_cartoes)

# 3. Agrupar cartões recebidos como time_1 e como time_2
amarelos_casa = wc.groupby('time_1')['num_cartoes_1'].sum().reset_index()
amarelos_casa.columns = ['país', 'amarelos']

amarelos_fora = wc.groupby('time_2')['num_cartoes_2'].sum().reset_index()
amarelos_fora.columns = ['país', 'amarelos']

# 4. Consolidar os dados de todos os países
cartoes_totais = pd.concat([amarelos_casa, amarelos_fora])
resultado = cartoes_totais.groupby('país')['amarelos'].sum().reset_index()

# 5. Formatação final
resultado.columns = ['país', 'cartões amarelos']
resultado = resultado.sort_values(by='cartões amarelos', ascending=False)

# Imprimir o Top 10 para conferência
print(resultado.head(10).to_string(index=False))
````

#### Q6.
Qual o top10 jogadores com mais gols em copa? Considere gols em jogo e gols de pênalti.

Para conseguir essa informação, você precisará trabalhar com as colunas:
```
10  gols_1_detalhes         
11  gols_2_detalhes    
14  gols_1_penalti     
15  gols_2_penalti     
```

Essas também são colunas textuais, onde cada observações apresenta todos os gols de uma partida separados pelo caracter `|`. Cada gol está na forma `'Jogador(a) · minuto’'`. Por exemplo:
```
'Alex Morgan · 12’|Rose Lavelle · 20’'
```

Lembre de levar em consideração que muitas observações são `NaN`. 

Recomendo criar um dicionário à parte, onde cada chave será um jogador encontrado nessas colunas do dataframe `wc` e o valor correspondente será a contagem de ocorrências desses nomes.

Em seguida basta converter o seu dicionário em um dataframe e imprimí-lo na forma (valores não são inventados 😁):
```
jogador(a)      num_gols 
Marta           17
Miroslav Klose  16
... 
```

````python
import pandas as pd

# 1. Inicializar o dicionário de artilharia
artilharia = {}

def processar_gols(celula):
    # Se for NaN ou nulo, ignoramos
    if pd.isna(celula) or celula == "" or celula == "[]":
        return
    
    # Remove colchetes e aspas se a string estiver formatada como lista
    limpo = str(celula).replace("[", "").replace("]", "").replace("'", "")
    
    # Os gols são separados por '|'
    gols_individuais = limpo.split('|')
    
    for registro in gols_individuais:
        if '·' in registro:
            # O formato é 'Jogador(a) · minuto’', então pegamos a parte antes do '·'
            nome = registro.split('·')[0].strip()
            
            # Contabiliza no dicionário
            artilharia[nome] = artilharia.get(nome, 0) + 1

# 2. Aplicar a função em todas as colunas de gols detalhados e pênaltis
colunas_gols = ['gols_1_detalhes', 'gols_2_detalhes', 'gols_1_penalti', 'gols_2_penalti']

for coluna in colunas_gols:
    wc[coluna].apply(processar_gols)

# 3. Converter o dicionário para um DataFrame
df_artilharia = pd.DataFrame(list(artilharia.items()), columns=['jogador(a)', 'num_gols'])

# 4. Ordenar e exibir o Top 10
df_artilharia = df_artilharia.sort_values(by='num_gols', ascending=False)

print(df_artilharia.head(10).to_string(index=False))
````
